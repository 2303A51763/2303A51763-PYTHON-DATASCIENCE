```python
import numpy as np
import matplotlib.pyplot as plt

temps_celsius = np.array([12, 15, 18, 24, 30, 35, 33, 32, 28, 22, 18, 14])

avg_temp = np.mean(temps_celsius)

hottest_month = np.argmax(temps_celsius) + 1
coldest_month = np.argmin(temps_celsius) + 1

norm_temps = (temps_celsius - np.min(temps_celsius)) / (np.max(temps_celsius) - np.min(temps_celsius))

months_above_avg = np.where(temps_celsius > avg_temp)[0] + 1

temps_fahrenheit = temps_celsius * 9/5 + 32

print("Monthly Temperatures (°C):", temps_celsius)
print("Average Temperature (°C):", round(avg_temp, 2))
print("Hottest Month:", hottest_month, "with", temps_celsius[hottest_month-1], "°C")
print("Coldest Month:", coldest_month, "with", temps_celsius[coldest_month-1], "°C")
print("Normalized Temperatures:", np.round(norm_temps, 2))
print("Months above average:", months_above_avg)
print("Temperatures in Fahrenheit (°F):", np.round(temps_fahrenheit, 1))

plt.plot(range(1, 13), temps_celsius, marker='o', label="Celsius")
plt.plot(range(1, 13), temps_fahrenheit, marker='s', label="Fahrenheit", linestyle="--")
plt.axhline(avg_temp, color="red", linestyle=":", label="Average (°C)")
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()

```
